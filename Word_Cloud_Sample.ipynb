```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
text = open('data.txt').read()
```

```python
mask_image = np.array(Image.open("r.png").convert('LA'))
```

```python
stopwords = set(STOPWORDS)
```

```python
wc = WordCloud(background_color = "white",max_words = 40000 ,mask = mask_image, stopwords=stopwords)
```

```python
wc.generate(text)
```

```python
wc.to_file('data.png')
```

```python
plt.imshow(wc, interpolation='bilinear',cmap = "hot")
plt.axis("off")
plt.figure()
plt.show()
```
